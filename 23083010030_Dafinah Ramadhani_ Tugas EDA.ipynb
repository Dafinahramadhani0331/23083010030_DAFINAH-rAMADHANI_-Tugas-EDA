## Amazon Fine Food Reviews
### Analyze 500,000 food reviews from amazon


Dataset ini terdiri dari ulasan makanan dari Amazon dan mencakup periode lebih dari 10 tahun, yaitu dari Oktober 1999 hingga Oktober 2012. Dataset ini mencakup sekitar 568.454 ulasan, dengan informasi mengenai produk, pengguna, peringkat, dan ulasan teks. Data juga mencakup ulasan dari semua kategori produk di Amazon.

Variabel dalam Dataset:

1. id: Identifikasi unik untuk setiap ulasan (numerik).
2. productid: Identifikasi unik untuk setiap produk contohnya 'B00813GRG4' (Kategorik).
3. userid: Identifikasi unik untuk setiap pengguna contohnya 'A1D87F6ZCVE5NK'(kategorik).
4. profile_name: Nama profil pengguna yang memberikan ulasan contohnya "Natalia Corres"(kategorik).
5. helpfulnessnumerator: Jumlah pengguna yang menganggap ulasan bermanfaat contohnya ' angka 1 sampai 5' disini ada 866 angka      (numerik)
6. helpfulnessdenominator: Jumlah pengguna yang menunjukan apakah mereka menganggap ulasan tersebut bermanfaat atau tidak       contohnya 'angka 1 sampai 5' disini ada 923 angka  (numerik).
7. score: Skor peringkat yang diberikan oleh pengguna, 'contohnya  antara peringkat 1 dam 5'  (numerik).
8. time:stempel waktu untuk peninjauan contohnya '1320364800' ada 939m sampai 1,35rb (numerik).
9. summary: Ringkasan singkat dari ulasan brisi nilai-nilai unik contohnya 'Harga lebih baik untuk ini di Target ' (teks).
10. text: Isi lengkap dari ulasan contohnya 'Saya membeli rasa mangga, dan menurut saya rasanya sama sekali tidak seperti  mangga. Tidak ada sedikit pun rasa manis...' (teks).

```python
import pandas as pd
import matplotlib.pyplot as plt  # Pastikan ini diimpor
import seaborn as sns

data = pd.read_csv("C://Users//ASUS//Documents//Reviews.csv")

score_stats = data['Score'].describe()
print("Descriptive Statistics for 'Score':")
print(score_stats)

plt.figure(figsize=(12, 6))
sns.histplot(data['Score'], bins=range(1, 7), kde=False, color='blue')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.xticks(range(1, 6))  # Assuming scores are between 1 and 5
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
sns.kdeplot(data['Score'], bw_adjust=0.5, fill=True, color='green')
plt.title('Density Plot of Scores')
plt.xlabel('Score')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

### Temporal Analysis:

#### Tren Waktu: Analisis tren ulasan dari waktu ke waktu untuk mengidentifikasi pola musiman atau perubahan dalam waktu.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C://Users//ASUS//Documents//Reviews.csv")

data['Date'] = pd.to_datetime(data['Time'], unit='s')

monthly_review_counts = data['Date'].dt.to_period('M').value_counts().sort_index()

plt.figure(figsize=(12, 6))
monthly_review_counts.plot(kind='line', color='blue', marker='o')
plt.title('Number of Reviews per Month')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

#### Perubahan Rating: Analisis bagaimana skor rata-rata berubah selama periode waktu.


```python

monthly_avg_score = data.groupby(data['Date'].dt.to_period('M'))['Score'].mean()

plt.figure(figsize=(12, 6))
monthly_avg_score.plot(kind='line', color='green', marker='o')
plt.title('Average Score per Month')
plt.xlabel('Month')
plt.ylabel('Average Score')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

### Analisis per produk

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("C://Users//ASUS//Documents//Reviews.csv")

# 1. Hitung jumlah ulasan per produk
product_review_counts = data['ProductId'].value_counts()

# 2. Hitung rata-rata skor per produk
product_avg_score = data.groupby('ProductId')['Score'].mean()

# 3. Gabungkan data jumlah ulasan dan rata-rata skor
product_analysis = pd.DataFrame({
    'review_count': product_review_counts,
    'average_score': product_avg_score
}).reset_index()

# 4. Ubah nama kolom untuk lebih jelas
product_analysis.columns = ['ProductId', 'ReviewCount', 'AverageScore']

# 5. Sortir berdasarkan jumlah ulasan untuk melihat produk paling populer
popular_products = product_analysis.sort_values(by='ReviewCount', ascending=False)

# 6. Sortir berdasarkan rata-rata skor untuk melihat produk dengan ulasan terbaik
top_rated_products = product_analysis.sort_values(by='AverageScore', ascending=False)

# 7. Plot produk paling populer berdasarkan jumlah ulasan
plt.figure(figsize=(12, 6))
sns.barplot(x='ReviewCount', y='ProductId', data=popular_products.head(10), palette='viridis')
plt.title('Top 10 Most Reviewed Products')
plt.xlabel('Number of Reviews')
plt.ylabel('Product Id')
plt.grid(True)
plt.show()

# 8. Plot produk dengan skor rata-rata tertinggi
plt.figure(figsize=(12, 6))
sns.barplot(x='AverageScore', y='ProductId', data=top_rated_products.head(10), palette='coolwarm')
plt.title('Top 10 Highest Rated Products')
plt.xlabel('Average Score')
plt.ylabel('Product Id')
plt.grid(True)
plt.show()
```

### Korelasi dan Hubungan:
#### Menganalisis korelasi antara variabel numerik seperti score, helpfulnessnumerator, dan helpfulnessdenominator & hubungan antara skor dan kualitas 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C://Users//ASUS//Documents//Reviews.csv")

# 1. Korelasi antara variabel numerik
correlation_matrix = data[['Score', 'HelpfulnessNumerator', 'HelpfulnessDenominator']].corr()

print("Correlation Matrix:")
print(correlation_matrix)

# Visualisasi korelasi dengan heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix between Score, HelpfulnessNumerator, and HelpfulnessDenominator')
plt.show()

# Menghitung rasio kegunaan (HelpfulnessNumerator / HelpfulnessDenominator)
data['HelpfulnessRatio'] = data['HelpfulnessNumerator'] / data['HelpfulnessDenominator']

# Plot scatter plot untuk melihat hubungan antara skor dan rasio kegunaan
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Score', y='HelpfulnessRatio', data=data, alpha=0.5)
plt.title('Relationship between Score and Helpfulness Ratio')
plt.xlabel('Score')
plt.ylabel('Helpfulness Ratio')
plt.grid(True)
plt.show()
```
